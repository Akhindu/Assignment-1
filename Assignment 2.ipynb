```python
# Question 1: Use yfinance to Extract Stock 
import yfinance as yf
import pandas as pd
tesla_data = yf.download('TSLA')
tesla_data.reset_index(inplace=True)
print(tesla_data.head())
```

```python
gme_data = yf.download('GME', start='2020-01-01', end='2023-01-01')
gme_data.reset_index(inplace=True)
gme_data.to_csv('gme_data.csv', index=False)
print(gme_data.head())
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from io import StringIO


url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue#:~:text=GameStop%20revenue%20for%20the%20twelve,a%2011.04%25%20decline%20from%202023.'  # Example URL


response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')


table = soup.find_all('table')[0]  


gme_revenue = pd.read_html(StringIO(str(table)))[0]


print(gme_revenue.tail())
```

```python
# Question 5: Plot Tesla Stock Graph
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
def make_graph(data):
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'], data['Close'], label='Tesla Stock Price', color='blue')
    plt.title('Tesla Stock Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()
    plt.tight_layout()  
    plt.show()
tesla_data = yf.download('TSLA', start='2020-01-01', end='2023-01-01')
tesla_data.reset_index(inplace=True)
make_graph(tesla_data)
```
